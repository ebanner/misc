# Get Lambda Names

```python
result = !aws lambda list-functions \
  --query 'Functions[].FunctionName' \
  --output text

lambda_names = result[0].split() + result[1].split()

lambda_names
```

# Get Lambda Configs

```python
import json

from tqdm import tqdm

lambdas = []
for lambda_name in tqdm(lambda_names):
    result = !aws lambda get-function-url-config \
      --function-name {lambda_name}

    try:
        result = json.loads('\n'.join(result))
    except:
        pass
    lambdas.append(result)

lambdas
```

# Extract Lambda URLs

```python
import pandas as pd

lambdas_filtered = []
for lambda_name, result in zip(lambda_names, lambdas):
    if type(result) != dict:
        continue

    result['lambda_name'] = lambda_name
    lambdas_filtered.append(result)

df = pd.DataFrame(lambdas_filtered)[['lambda_name', 'FunctionUrl']]

df
```
